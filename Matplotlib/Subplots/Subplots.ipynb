# Subplots

<hr>

<b><i>If we want to create plots in more plots or work in more object oriented manner then it is best create our plot with subplot method</i></b>

<hr>

```python
# Importing modules
import pandas as pd
from matplotlib import pyplot as plt
```

<hr>

```python
# Getting the data
data = pd.read_csv('data.csv')
```

<hr>

### Non Object Oriented Manner

```python
plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.legend()

plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Ages")

plt.tight_layout()

plt.show()
```

<hr>

<b><i>
Every time we are doing something with our plot we are using the plt object, we never create that object
<br>
In the above plot we have single figure and single axis
<br>
The figure is the container holding our plots and axis are the actual plot
<br>
A figure can have multiple plots
To do this we will use Subplot method
</b></i>

<hr>

```python
plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Intantiating Figure and Axes(Plots)
fig, ax = plt.subplots()
# By default plot(ax) is set to 1 because by default subplot create a figure and then specify a certain number of rows and columns of axes
# So if we don't pass in our numbers of rows and columns then it defaults to 1 by 1 i.e. 1 row and 1 column which is simply 1 axis

# After intantiating our figure and axis we have to replace all plt with ax

ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')
ax.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

ax.legend()

# When it comes to our title and labels we add a prefix of set_ to methods
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")
ax.set_title("Median Salary (USD) by Ages")

# tight_layout and show methods are methods of pyplot object so changes need to be done there
plt.tight_layout()
plt.show()
```

<hr>

### Making subplots

```python
plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Ages")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.set_title("Median Salary (USD) by Ages")

plt.tight_layout()
plt.show()
```

<hr>

### Refactoring our plots

```python
plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Ages")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

<hr>

```python
# If we want x tick marks only on bottom then that can be done by passing an argument to the subplot and that is "sharex" and set that to True

# There is also "sharey" argument
# "sharex" will only label bottom ticks or two rows and "sharey" will only label the left most ticks or two columns

plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Ages")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

<hr>

### Having two plots on different figures instead of having them in same plot

```python
# To do that we need to create another figure and axis

plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1= plt.subplots()
fig2, ax2= plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Ages")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.set_title("Median Salary (USD) by Ages")

plt.tight_layout()
plt.show()
```

<hr>

### Saving different figures

```python
# To do that we need to create another figure and axis

plt.style.use('seaborn')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1= plt.subplots()
fig2, ax2= plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Ages")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.set_title("Median Salary (USD) by Ages")

plt.tight_layout()
plt.show()

# Saving different figures
fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
