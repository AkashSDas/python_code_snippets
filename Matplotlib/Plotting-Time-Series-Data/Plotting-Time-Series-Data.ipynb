# Plotting Time Series Data

<hr>

```python
# Importing modules
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
```

<hr>

### Basic Example of Plotting Time Series Data

```python
plt.style.use('seaborn')

dates = [
  datetime(2019, 5, 24),
  datetime(2019, 5, 25),
  datetime(2019, 5, 26),
  datetime(2019, 5, 27),
  datetime(2019, 5, 28),
  datetime(2019, 5, 29),
  datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y)

plt.tight_layout()

plt.show()
```

```python
# To have plot that have connected lines instead of markers
# We can do that by setting value of 'linestyle' as 'solid'

plt.style.use('seaborn')

dates = [
  datetime(2019, 5, 24),
  datetime(2019, 5, 25),
  datetime(2019, 5, 26),
  datetime(2019, 5, 27),
  datetime(2019, 5, 28),
  datetime(2019, 5, 29),
  datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.tight_layout()

plt.show()
```

```python
# If we don't want markers then we can set 'marker' to None

plt.style.use('seaborn')

dates = [
  datetime(2019, 5, 24),
  datetime(2019, 5, 25),
  datetime(2019, 5, 26),
  datetime(2019, 5, 27),
  datetime(2019, 5, 28),
  datetime(2019, 5, 29),
  datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid', marker=None)

plt.tight_layout()

plt.show()
```

### Rotating the dates in X-axis

```python
# To do that 

plt.style.use('seaborn')

dates = [
  datetime(2019, 5, 24),
  datetime(2019, 5, 25),
  datetime(2019, 5, 26),
  datetime(2019, 5, 27),
  datetime(2019, 5, 28),
  datetime(2019, 5, 29),
  datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

# To get current figure from pyplot we will use gcf(get current figure) method
plt.gcf().autofmt_xdate()
# Applying the auto format of dates in x-axis

# autofmt_xdate method is on our figure and not on the pyplot object that we are using

plt.tight_layout()

plt.show()
```

<hr>

### Formatting the dates

```python
plt.style.use('seaborn')

dates = [
  datetime(2019, 5, 24),
  datetime(2019, 5, 25),
  datetime(2019, 5, 26),
  datetime(2019, 5, 27),
  datetime(2019, 5, 28),
  datetime(2019, 5, 29),
  datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

# Getting a date format for dates in x-axis
date_format = mpl_dates.DateFormatter('%b, %d, %Y')

# To set the date format we need to grab the axis just like we did for the figure 
# For this we will use the gca(get current axis) method
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()

plt.show()
```

### Working with real data

```python
# Getting data
data = pd.read_csv('data.csv')
```

```python
plt.style.use('seaborn')

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

<i>In the above plot our x-axis looks like we have plotted dates but those are strings</i><br>
<i>Thats why we have 2019-05-17 at the end of the x-axis since that is the last string in our csv file</i>

<hr>

```python
plt.style.use('seaborn')

data['Date'] = pd.to_datetime(data['Date'])

# Sorting the data as per dates
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
