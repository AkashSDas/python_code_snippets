# Scatter Plots
<hr>

<i>Scatter Plots are great when we want to show relationship between two sets of values and see if they are correlated</i>
<hr>

```python
# Importing modules
import pandas as pd
from matplotlib import pyplot as plt
```

<hr>

### Basic Example of Scatter Plot

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y)

plt.tight_layout()

plt.show()
```

<hr>

### Changing size of our data-points

```python
# For this we are using the "s" parameter of our scatter method

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100)

plt.tight_layout()

plt.show()
```

<hr>

### Changing the color and marker style of our data-points

```python
# For this we are using the "c" and "marker" parameters respectively

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c="green", marker="X")

plt.tight_layout()

plt.show()
```

<hr>

### Adding edges and alpha to our data-points

```python
# For this we are using the  "edgecolor", "linewidth" and "alpha parameter"

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c="green", edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()

plt.show()
```

<hr>

### Colors and Sizes on per mark basis

<p>
    <i>Why do we want multiple colors and sizes?</i><br>
    <i>Having multiple colors and sizes allow us to add multiple datasets to our plot.</i>
</p>

<p>
    <b>Lets say our data represent a survey result and we want to add some extra data like there rating(from 1 to 10) and we want to plot there ratings also.</b><br>
    <b>In such use cases having multiple colors and sizes are very useful.</b>
</p>

<hr>

### Changing colors

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# Setting "c" parameter to colors list which represent the ratings
plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75)
# Different colors of data-point in the plot represent different values in the colors list

plt.tight_layout()

plt.show()
```

<hr>

### Using color maps

```python
# For this we will be using the "cmap" parameter in our scatter method

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()

plt.show()
```

<i>There are tons of cmaps</i>

<hr>

### Adding a label to our color maps

```python
# We can add label to our color maps so that people viewing our plot can kown what each value specific
# For this we will add a color map legend

plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)

# Setting label for our color maps
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()

plt.show()
```

<hr>

### Changing sizes

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# Adding sizes list
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

# Putting value of "s" as sizes list
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()

plt.show()
```

### Working with real data

```python
# Getting the data
data = pd.read_csv('data.csv')
```

```python
plt.style.use('seaborn')

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

<i>Our plot is cluttered in lower left due outliers. To lessen the skewness due to those outliers we will use log</i>

```python
plt.style.use('seaborn')

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

# To lessen how much outliner skew the plot
plt.xscale('log')
plt.yscale('log')
# We are using log scale instead of regualar scale

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

<hr>

### Using cmps to understand the ratio more clearly

```python
plt.style.use('seaborn')

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
