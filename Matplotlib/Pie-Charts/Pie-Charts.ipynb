# Pie Charts
<hr>

```python
# Importing pyplot from matplotlib module
from matplotlib import pyplot as plt
```

<hr>

### Basic Example of Pie Chart

```python
plt.style.use("fivethirtyeight")

slices = [60, 40]

plt.pie(slices)

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### Adding labels to our pie chart

```python
plt.style.use("fivethirtyeight")

slices = [60, 40]
labels = ["Sixty", "Fourty"]

plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

```python
# We don't need to worry about our values getting added to hundres only we can take whatever value we want and python will take care of it 

plt.style.use("fivethirtyeight")

slices = [120, 80]
labels = ["Sixty", "Fourty"]

plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### Getting separators near the edges

```python
# For this we have to pass the wedgeproperties

plt.style.use("fivethirtyeight")

slices = [120, 80]
labels = ["Sixty", "Fourty"]

plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### Specifying colors for our slices

```python
plt.style.use("fivethirtyeight")

slices = [120, 80, 30, 20]
labels = ["Sixty", "Fourty", "Thirty", "Twenty"]
colors = ["blue", "red", "yellow", "green"]

plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### Using hex-colors for our slices

```python
plt.style.use("fivethirtyeight")

slices = [120, 80, 30, 20]
labels = ["Sixty", "Fourty", "Thirty", "Twenty"]
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

## Working with real data

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

```python
# Striping our data to just five languages

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### In our pie charts having emphasis on one part

```python
# A lot of times we want to emphasis on only one part of our Pie Chart
# Let's say we want are preparing for a talk on Python and we using a pie chart then we have to emphasis on the Python part of our pie chart
# To do the emphasis on certain parts of our pie chart we have to pass in explode argument and this will be a list of values which will offset the slices

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# Exploding python which is 4th item of our list
explode = [0, 0, 0, 0.1, 0]
# 0 will keep the slice as it is and any number beside 0 will represent the fraction of radius that we want to explode the value

plt.pie(slices, labels=labels, explode=explode, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

```python
# In explode any number beside 0 will represent the fraction of radius that we want to explode the value
# In the above cell it is 0.1 i.e. 10% of the radius 

# Here we will explode the Python slice by 0.5 i.e. 50% of the radius i.e. half the radius

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.5, 0]

plt.pie(slices, labels=labels, explode=explode, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### Adding shadow to our chart

```python
# To add shadow to our chart we have to pass shadow=True in pie method

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### startangle

```python
# If we want to control on how the chart is rotated we can pass the startangle argument

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```

<hr>

### autopct

```python
# To display the percentage of how much each slice is taking in our pie chart

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%', wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()

plt.show()
```
