# Plot Fills

## Filling Area on Line Plots

<hr>

```python
# Importing modules
import pandas as pd
from matplotlib import pyplot as plt
```

<hr>

```python
# Getting the data
data = pd.read_csv("data.csv")
```

<hr>

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

<hr>

### Adding Fills to our Plot

```python
# Filling the entire plot under the Python plotted data

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries)
# fill_between method has first parameter as x-aixs, second one is what we want to fill between
# Third parameter can be passed like we can say we want to fill between py_salaries(y1) and dev_salaries(y2) but by default the y2 is set to 0

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

```python
# Putting an alpha to make it look better

# Filling the entire plot under the Python plotted data

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

<hr>

### Condition Thresholds

```python
# Condition Thresholds can be useful to get some feedback from the data that we have

# Our fill is defaulting to 0 i.e. it is filling between our y-values and 0 (all the way to the bottom) but we can change that

# Let the median salary be 57287
overall_median = 57287

# Lets fill the plot between overall_median and py_salaries instead of 0 and py_salaries
# To do that we can pass it has the third argument in our fill_between method

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

<hr>

### Giving additional conditions so that whenever those conditions are meet, plot will be filled

```python
# To do the above task we have to add a "where" parameter in our fill_between method

overall_median = 57287

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)
# interpolate ==> it makes sure that certain x inetersection don't get cliped and all of the region gets filled correctly

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

```python
# To fill area below overall_median we have to use another fill_between method

overall_median = 57287

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, alpha=0.25)

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

<hr>

### To give custom colors to our fills

```python
# Lets do something like when we are above median salary our plot is filled with green and when we are below median salary our plot is filled with red
# For this we have to use the "color" parameter in our fill_between method

overall_median = 57287

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), color='red', interpolate=True, alpha=0.25)

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

<hr>

### To fill in between two plots

```python
# Lets say we want to fill in between Python salaries and our Developer salaries by age

# So instead of using the overall_median we will pass dev_salaries as third argument in the fill_between

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), color='red', interpolate=True, alpha=0.25)

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.show()
```

<hr>

### Label the sections of fill plot so anyone who sees our plot can understand them easily

```python
# To do that we have pass in a label in our fill_between method

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label="Python")

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), color='red', interpolate=True, alpha=0.25, label='Below Avg')

plt.legend()

plt.title("Median Salaary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.show()
```
